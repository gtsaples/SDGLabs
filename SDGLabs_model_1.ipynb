# Bass Model of Diffusion - Part 1

One of the most famous models that has been used to describe the adoption of new products and services is the one by Bass (Bass, 1969). The Bass Innovation model is based on the idea that these new products and services can spread through a potential market in a viral way (Lai and Poltera, 2009).

The model is using past data to determine the size of the market (Lee and Oh, 2020) and uses the notion of "word of mouth processes" to show how the new product can be adopted (Lai and Poltera, 2009). Consequently, in its basic form the model accounts for only two aspects: existing users of the product or service and potential users.

Even the most basic form of the model can reveal an insight that can elude many managers of organizations and businesses: growth cannot be eternal and continuous. Since the market is limited, the number of clients will also be limited.
